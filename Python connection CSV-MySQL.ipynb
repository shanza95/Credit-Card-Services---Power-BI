```python
# connect to mysql db
import mysql.connector
from mysql.connector import Error
 
try:
    # Establish the connection
 
    connection = mysql.connector.connect(
        host='127.0.0.1',         # or your host, e.g., '127.0.0.1'
        user='root',     # your MySQL username
        password='root', # your MySQL password
        database='cc_db',  # the database you want to connect to
        allow_local_infile=True
    )
    
    cursor = connection.cursor()
    if connection.is_connected():
        db_Info = connection.get_server_info()
        print(" Connected to MySQL Server version", db_Info)
        cursor = connection.cursor()
        cursor.execute("SELECT DATABASE();")
        record = cursor.fetchone()
        print(" You're connected to database:", record)
 
except Error as e:
    print("Error while connecting to MySQL", e)
 
finally:
    if 'connection' in locals() and connection.is_connected():
        cursor.close()
        connection.close()
        print("MySQL connection is closed")
```

```python
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="cc_db",
    allow_local_infile=True
)

cursor = conn.cursor()

# Function to import a CSV into a specified table
def import_csv(file_name, table_name):
    query = f"""
    LOAD DATA LOCAL INFILE 'C:/Users/Amir/Dropbox/web projects/credit card project -POWER BI/Datasets/{file_name}'
    INTO TABLE {table_name}
    FIELDS TERMINATED BY ','
    ENCLOSED BY '"'
    LINES TERMINATED BY '\\n'
    IGNORE 1 LINES;
"""

    cursor.execute(query)
    conn.commit()
    print(f"CSV '{file_name}' imported into table '{table_name}' successfully!")

# import CSV file customer.csv into cust_detail
import_csv("customer.csv", "cust_detail")
# import CSV file credit_card into cc_detail
import_csv("credit_card.csv", "cc_detail")

# close connection
cursor.close()
conn.close()

print("CSV imported using LOAD DATA LOCAL INFILE!")
```
